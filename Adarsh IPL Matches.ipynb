```python
#PLAYERS OF THE MATCH(WHICH PLAYER WON HOW MUCH TIME)
man_df=df.groupby("player_of_match")[["id"]].count()
man_df=man_df.sort_values("id",ascending=False).head(10)
man_df.rename(columns={"id":"No_of_Awards"},inplace=True)
man_df                                     
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
plt.figure(figsize=(15,5))
sns.barplot(y=man_df.index,x="No_of_Awards",data=man_df)
plt.title("MAN OF THE MATCH PLAYERS")
```

```python
#IN WHICH CITIES MATCHES ARE WON MOST
city_win=df.groupby("city")[["id"]].count()
city_match_win=city_win.sort_values("id",ascending=False).reset_index().head(1)
city_match_win
```

```python
#MAX MATCH WON BY ANY TEAM
team_win=df.groupby("winner")[["id"]].count()
team_win_match=team_win.sort_values("id",ascending=False).reset_index().head(1)
team_win_match.rename(columns={"id":"won_the_matches","winner":"Team"},inplace=True)
team_win_match
```

```python
#MAX MATCH WON BY ANY TEAM
team_win=df.groupby("winner")[["id"]].count()
team_win_match=team_win.sort_values("id",ascending=False).reset_index().head(10)
team_win_match.rename(columns={"id":"won_the_matches","winner":"Team"},inplace=True)
team_win_match
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
sns.barplot("won_the_matches","Team",data=team_win_match)
plt.title("EACH TEAM WINNING MATCHES")
```
